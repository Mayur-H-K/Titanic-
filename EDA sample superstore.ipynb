# Visualize the outliers 

```python
print(retail.Sales.describe([.1,.2,.3,.4,.5,.6,.7,.8,.9,.95, .99]))
plt.figure(figsize=(6,6))
retail.Sales.plot(kind='box')
```

Sales have a huge variations since there are many types of categories and sub-categories which have variation in their price.

```python
print(retail.Quantity.describe([.1,.2,.3,.4,.5,.6,.7,.8,.9,.95, .99]))
plt.figure(figsize=(6,6))
retail.Quantity.plot(kind='box')
```

```python
print(retail.Discount.describe([.1,.2,.3,.4,.5,.6,.7,.8,.9,.95, .99]))
plt.figure(figsize=(6,6))
retail.Discount.plot(kind='box')
```

```python
print(retail.Profit.describe([.1,.2,.3,.4,.5,.6,.7,.8,.9,.95, .99]))
plt.figure(figsize=(6,6))
retail.Profit.plot(kind='box')
```

Profit varies from positive to negative which indicates there are profits and losses

```python
#Lets see how sub-categories are distributed wrt to category
plt.figure(figsize=(16,8))
sns.countplot('Category',data=retail,color='b')
plt.show()
```

```python
plt.figure(figsize=(12,10))
retail['Sub-Category'].value_counts().plot.pie(autopct="%1.1f%%")
plt.show()
```

```python
plt.figure(figsize=(15,8))
retail.groupby('Sub-Category')['Profit','Sales'].agg(['sum']).plot.bar()
plt.title('Total Profit and Sales per Sub-Category')
# plt.legend('Profit')
# plt.legend('Sales')
plt.show()
```

```python
plt.figure(figsize=(15,8))
sns.countplot(x="Category", hue="Region", data=retail)
plt.show()
```

```python
plt.figure(figsize=(15,8))
sns.countplot(x="Sub-Category", hue="Region", data=retail)
plt.show()
```

```python
plt.figure(figsize=(15,8))
sns.countplot(x="Region", hue="Ship Mode", data=retail)
plt.show()
```

```python
plt.figure(figsize=(15,8))
sns.countplot(x="Sub-Category", hue="Segment", data=retail)
plt.show()
```

```python
plt.figure(figsize=(15,8))
sns.heatmap(retail.corr(),cmap ="YlGnBu",annot = True)
plt.show()
```
